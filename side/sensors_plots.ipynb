```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Consts
datapath = '../rawdata/sensors/'
filename = datapath + 'nodes.txt'

f = open(filename)
sensors = {}

for line in f:
    s, lat, lon = line.split(' ')
    
    
    # loading sensor data
    sensorData = pd.read_pickle(datapath +s+ '.pkl')
    
    sensorData['dbas'] = sensorData['sum'] / sensorData['count']
    sensorData = sensorData[['dbas']]
    
    
    plt.figure(figsize=(15,5))
    plt.plot(sensorData['dbas'])
    plt.title(s)
    plt.show()
```
